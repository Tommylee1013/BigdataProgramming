### 실습 : pie chart 그리기
- slice labels
- auto-labeling the percentage (autopct)
- offsetting a slice with 'explode'
- drop shadow
- custom start angle : default is 0
- counter - clockwise

```python
import matplotlib.pyplot as plt
data = [26.2, 30.6, 16.2, 26.9]
label = ['Java','Ruby','C++','Python']
explode = (0, 0, 0, 0.1)
plt.style.use('default')
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(data, labels = label, shadow = True, startangle = 140, counterclock = False, 
        autopct = '%1.1f%%', explode = explode, colors = colors)
plt.show()
```
