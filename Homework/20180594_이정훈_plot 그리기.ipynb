### 실습 2
- 아래와 같은 Cosine Graph를 그리시오
- 틱 라벨 문자열에서 \$$ 사이에 LaTeX수학 문자식을 넣어 해결하시오

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', '0', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,1])
plt.show()
```

### 실습 3
- sine 함수와 cosine함수를 아래와 같이 겹쳐 그리시오

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
S = np.sin(X)
plt.plot(X, C)
plt.plot(X, S)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', '0', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,1])
plt.show()
```
