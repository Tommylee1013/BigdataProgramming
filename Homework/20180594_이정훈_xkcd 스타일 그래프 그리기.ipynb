### 실습 4
- 아래와 같이 $x = \frac{2\pi}{3}$위치에 annontation을 넣은 cosine함수를 그리시오

```python
import matplotlib.pyplot as plt, numpy as np, pandas as pd
```

```python
plt.xkcd()
plt.figure(figsize = (6,4))
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(2 * np.pi/3, np.cos(2 * np.pi/3), c = 'red', lw = 5, marker = 'o', ms = 8)
plt.plot(X, C)
plt.annotate(r'-0.5 Here!', xy = (2 * np.pi/3, np.cos(2 * np.pi/3)), xytext = (-85,-50), xycoords = 'data', 
             textcoords = 'offset points', fontsize = 16, 
             arrowprops = dict(arrowstyle = '->'))
plt.title('XKCD style plot!!!')
plt.show()
```
