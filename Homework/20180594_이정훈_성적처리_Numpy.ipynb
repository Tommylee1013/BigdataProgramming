### 과제 10 : 성적처리 Numpy 이용하기

```python
import numpy as np
import matplotlib.pyplot as plt

# Set Database
grade = np.array([[75,89,92,78],
                  [90,76,88,83],
                  [55,99,81,88],
                  [80,67,75,91]])

# Label
name = np.array(['Lee','Park','Jung','Choi'])
subject = np.array(['kor','Math','Eng','Com'])
subject_kor = ['국어','수학','영어','컴']

# Define function

def print_grade(data, subject) :
    print('총점 :', grade.sum(axis = 0).mean())
    print('각 과목의 평균과 최고점')
    for i in range(len(subject)) :
        print(subject[i],':',data.mean(axis = 0)[i],',',grade.max(axis = 0)[i])

def plot_grade(data, subject, name) :
    # plot 1 
    plt.style.use('seaborn')
    plt.figure(figsize = (6,4))
    plt.plot(name, data.mean(axis = 1))
    plt.xlabel('Student', fontsize = 12, fontfamily = 'Serif')
    plt.ylabel('Average score', fontsize = 12, fontfamily = 'Serif')
    plt.title('Average score of Student', fontsize = 15, fontfamily = 'Serif')
    plt.show()
    
    # plot 2
    plt.figure(figsize = (6,4))
    plt.plot(name, data.mean(axis = 0), label = 'subject avg')
    plt.plot(name, data.max(axis = 0), label = 'subject max')
    plt.xlabel('Subjects', fontsize = 12, fontfamily = 'Serif')
    plt.ylabel('Score', fontsize = 12, fontfamily = 'Serif')
    plt.legend()
    plt.title('Average and Top score of Subjects', fontsize = 15, fontfamily = 'Serif')
    plt.show()
    
def main() :
    print_grade(grade, subject)
    plot_grade(grade, subject_kor, name)
    
# Run program
if __name__ == '__main__' :
    main()
```
